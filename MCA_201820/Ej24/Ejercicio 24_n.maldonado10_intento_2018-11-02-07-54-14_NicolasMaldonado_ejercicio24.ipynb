```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
data = pd.read_csv('dengue.csv')
data = data.dropna()

X = data[['humid', 'temp' , 'h10pix', 'trees' , 'Xmin', 'Ymin']]
X = np.array(X)
```

```python
fig = plt.figure(figsize=[20,25])
n = 1
for i in range(6):
    for j in range(i+1,6):
        sub = plt.subplot(5,3,n)
        plt.scatter(X[:,i],X[:,j])
        n += 1
```

```python
num = np.arange(1,10)
inertia = np.zeros(len(num))

for i in num:
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit_transform(X)

    inertia[i-1] = kmeans.inertia_
    
plt.plot(num,inertia,marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
```

```python
print('El nmero ptimo es entonces 4 y vemos para K=4:')

kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit_transform(X)

fig = plt.figure(figsize=[20,25])
n = 1
for i in range(6):
    for j in range(i+1,6):
        sub = plt.subplot(5,3,n)
        plt.scatter(X[:,i],X[:,j],c=kmeans.labels_,cmap='rainbow')
        n += 1
```
