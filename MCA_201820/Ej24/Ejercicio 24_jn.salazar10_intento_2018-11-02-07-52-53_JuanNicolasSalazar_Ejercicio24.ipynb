```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

### Parte 1

```python
data1 = pd.read_csv('dengue.csv')
data1.dropna(inplace = True)
X1 = data1[['humid', 'temp', 'h10pix', 'trees' , 'Xmin', 'Ymin']].values
```

```python
inertia = []
for i in range(10):
    km = KMeans(n_clusters = i+1)
    km.fit(X1)
    inertia.append(km.inertia_)
```

```python
plt.plot(range(1,11), inertia)
plt.scatter(range(1,11), inertia)
plt.xlabel('# de clusters')
plt.ylabel('inertia')
print('De la gráfica se observa que a partir de n_clusters > 3, la distancia de los centroides a los datos \nno disminuye significativamente, luego el óptimo es 3 clusters')
```
